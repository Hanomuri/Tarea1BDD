## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql
CREATE TABLE director(
    id_director INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INT NOT NULL
);
```

### Tabla "Actor"

```python
%%sql
CREATE TABLE actor(
    id_actor INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INT NOT NULL
);
```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula(
    id_pelicula INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    id_director INT,
    nombre VARCHAR(45) NOT NULL,
    genero VARCHAR(45) NOT NULL,
    duracion INT NOT NULL,
    fecha_publicacion DATE NOT NULL,
    FOREIGN KEY (id_director) REFERENCES director (id_director)
);
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula(
    id_actor INT REFERENCES actor,
    id_pelicula INT REFERENCES pelicula,
    PRIMARY KEY (id_actor, id_pelicula)
);
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director(
    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    id_director INT,
    anio INT NOT NULL,
    FOREIGN KEY (id_director) REFERENCES director (id_director)
);
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor(
    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    id_actor INT,
    anio INT NOT NULL,
    FOREIGN KEY (id_actor) REFERENCES actor (id_actor)
);
```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula(
    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    id_pelicula INT,
    anio INT NOT NULL,
    FOREIGN KEY (id_pelicula) REFERENCES pelicula (id_pelicula)
);
```

### Consulta 3
**Nombres de todos los actores que participaron en una película de género comedia.**


**Ejemplo:**
|Nombre       |
|-------------|
|Margot Robbie|
|...          |


**Solución:** Esta consulta identifica los nombres sin repetir de actores que han participado en películas del género comedia. Utiliza SELECT DISTINCT para evitar duplicados y combina dos operaciones de JOIN: una que enlaza a los actores con las películas en las que han participado y otra para vincular cada participación del actor con una película específica. Finalmente e filtra por el género comedia.

```python
%%sql
SELECT DISTINCT actor.nombre
FROM actor
JOIN actor_participa_en_pelicula ON actor.id_actor = actor_participa_en_pelicula.id_actor
JOIN pelicula ON actor_participa_en_pelicula.id_pelicula = pelicula.id_pelicula
WHERE pelicula.genero = 'Comedia';
```

### Consulta 4
**Nombre y cantidad de premios ganados por cada género, ordenados de mayor a menor según la cantidad de premios.**


**Ejemplo:**
|Género  |Cantidad de premios|
|--------|-------------------|
|Drama   |99                 |
|Fantasía|29                 |
|...     |...                |



**Solución:** Esta consulta calcula y lista la cantidad de premios obtenidos por películas de cada género. Realiza un `LEFT JOIN` entre las tablas `pelicula` y `premio_mejor_pelicula` para incluir todos los géneros, incluso aquellos sin premios. Los resultados se agrupan por género y se ordenan en función del número de premios recibidos, mostrando desde los géneros más premiados hasta los menos premiados.

```python
%%sql
SELECT pelicula.genero, COUNT(premio_mejor_pelicula.id_premio) AS cantidad_de_premios
FROM pelicula
LEFT JOIN premio_mejor_pelicula ON pelicula.id_pelicula = premio_mejor_pelicula.id_pelicula
GROUP BY pelicula.genero
ORDER BY cantidad_de_premios DESC;
```

### Consulta 5
**Mostrar peliculas que tengan 2 o mas actores/actrices ordenadas de mayor a menor segun la cantidad de actores/actrices.**


**Ejemplo:**
|Nombre     |Cantidad de actores|
|-----------|-------------------|
|Oppenheimer|7                  |
|Poor Things|4                  |
|...        |...                |


**Solución:** Esta consulta cuenta y lista las películas que tienen dos o más actores. Usa un `JOIN` para conectar las tablas `pelicula` y `actor_participa_en_pelicula`, relacionando películas con sus actores. Se agrupa por el nombre de la película y se usa la cláusula `HAVING` para filtrar y mostrar solo aquellas películas con al menos dos actores. Los resultados se ordenan de mayor a menor según la cantidad de actores, permitiendo ver qué películas han involucrado a más actores en su producción.

```python
%%sql

SELECT pelicula.nombre, COUNT(actor_participa_en_pelicula.id_actor) AS Cantidad_de_actores
FROM pelicula
JOIN actor_participa_en_pelicula ON pelicula.id_pelicula = actor_participa_en_pelicula.id_pelicula
GROUP BY pelicula.nombre
HAVING COUNT(actor_participa_en_pelicula.id_actor) >= 2
ORDER BY cantidad_de_actores DESC;
```

### Consulta 6
**Nombre de la película con la mayor brecha de edad entre el actor más joven y el actor más viejo que participen en ella, y las edades de dichos actores.** (Si hay empate, mostrar solamente uno).


**Ejemplo:**
|Nombre|Edad del actor más joven|Edad del actor más viejo|
|------|------------------------|------------------------|
|Barbie|16                      |92                      |


**Solución:**  Se realiza un `JOIN` entre las tablas `pelicula`, `actor_participa_en_pelicula`, y `actor` para vincular películas con los actores que aparecen en ellas. La consulta agrupa los resultados por el nombre de la película y calcula la edad mínima y máxima de los actores en cada película. El resultado se ordena por la diferencia entre la edad máxima y mínima, mostrando primero la película con la mayor brecha de edad. Para mostrar solo un dato se usa `LIMIT 1`.

```python
%%sql
SELECT 
    pelicula.nombre, 
    MIN(actor.edad) AS Edad_del_actor_más_joven,
    MAX(actor.edad) AS Edad_del_actor_más_viejo
    
FROM 
    pelicula
JOIN 
    actor_participa_en_pelicula ON pelicula.id_pelicula = actor_participa_en_pelicula.id_pelicula
JOIN 
    actor ON actor_participa_en_pelicula.id_actor = actor.id_actor
GROUP BY 
    pelicula.nombre
ORDER BY 
    MAX(actor.edad) - MIN(actor.edad) desc
LIMIT 1;
```

### Consulta 7
**Año de cada premio ganado por la película de mayor duración.** (Si hay empate, elegir solamente uno).


**Ejemplo:**

|2024|

**Solución:** Primero, se realiza un `JOIN` entre las tablas `premio_mejor_pelicula` y `pelicula` para asociar cada premio con la película correspondiente. Luego, usando `WHERE`, se filtran estos resultados para incluir solo aquellos que pertenecen a la película con la máxima duración, determinada por una subconsulta que calcula `MAX(duracion)` de todas las películas en la tabla `pelicula`. **Nombre de todos los actores que han trabajado con más de un director, la cantidad de directores con las que han trabajado y el nombre de una de las películas.**

|Nombre del actor|Cantidad de directores|Nombre de una película|

**Solución:** La consulta une las tablas `actor`, `actor_participa_en_pelicula`, y `pelicula` para relacionar a cada actor con las películas en las que ha participado y sus directores. Agrupa los resultados por el nombre del actor y utiliza `HAVING` para filtrar aquellos actores que han colaborado con más de un director.

```python
%%sql
SELECT actor.nombre AS nombre_del_actor, COUNT(DISTINCT pelicula.id_director) AS cantidad_directores, MIN(pelicula.nombre) AS nombre_de_una_pelicula
FROM actor
JOIN actor_participa_en_pelicula ON actor.id_actor = actor_participa_en_pelicula.id_actor
JOIN pelicula ON actor_participa_en_pelicula.id_pelicula = pelicula.id_pelicula
GROUP BY actor.nombre
HAVING COUNT(DISTINCT pelicula.id_director) > 1;
```

**Solución:** La consulta comienza uniendo la tabla `actor` con `actor_participa_en_pelicula` para relacionar a los actores con las películas en las que han actuado, y luego se une esta combinación con la tabla `pelicula` para incluir la información del director de cada película. Los resultados se agrupan por el nombre del actor (`GROUP BY actor.nombre`) y se utiliza `HAVING` para asegurarse de que cada actor listado solo ha trabajado con ese director específico (`COUNT(DISTINCT pelicula.id_director) = 1`).

```python
%%sql
SELECT actor.nombre
FROM actor
JOIN actor_participa_en_pelicula ON actor.id_actor = actor_participa_en_pelicula.id_actor
JOIN pelicula ON actor_participa_en_pelicula.id_pelicula = pelicula.id_pelicula
WHERE pelicula.id_director = 1
GROUP BY actor.nombre
HAVING COUNT(DISTINCT pelicula.id_director) = 1;
```

### Consulta BONUS

**Nombre de todos los actores que nunca han ganado un premio, todas las películas en las que han participado y el nombre, número total de películas y número total de premios de los directores de cada película.**

|Saoirse Ronan   |Little Women         |Greta Gerwig       |20                              |1                      |

**Solución:** Esta consulta comienza vinculando la tabla `actor` con `actor_participa_en_pelicula` para relacionar actores con sus películas correspondientes. Luego, se une esta información con la tabla `pelicula`, y posteriormente con `director` para obtener el nombre del director de cada película. Luego se utiliza un `LEFT JOIN` con una subconsulta que agrupa información sobre cada director. Esta subconsulta calcula el total de películas dirigidas por cada director y el total de premios que han ganado, utilizando las tablas `pelicula` y `premio_mejor_director`. Esto se logra mediante un `LEFT JOIN` en `pelicula` y `premio_mejor_director` para cada director, y luego se agrupa por `id_director`. La consulta filtra para incluir solo aquellos actores que nunca han ganado un premio (`premio_mejor_actor.id_premio IS NULL`).

```python
%%sql
SELECT 
    actor.nombre AS nombre_del_Actor,
    pelicula.nombre AS Nombre_de_la_pelicula,
    director.nombre AS Nombre_del_director,
    sub_director.total_peliculas AS Número_de_películas_del_director,
    sub_director.total_premios AS Número_de_premios_del_director
FROM 
    actor
JOIN 
    actor_participa_en_pelicula ON actor.id_actor = actor_participa_en_pelicula.id_actor
JOIN 
    pelicula ON actor_participa_en_pelicula.id_pelicula = pelicula.id_pelicula
JOIN 
    director ON pelicula.id_director = director.id_director
LEFT JOIN 
    premio_mejor_actor ON actor.id_actor = premio_mejor_actor.id_actor
LEFT JOIN 
    (SELECT 
         director.id_director,
         COUNT(DISTINCT pelicula.id_pelicula) AS total_peliculas,
         COUNT(premio_mejor_director.id_premio) AS total_premios
     FROM 
         director
     LEFT JOIN 
         pelicula ON director.id_director = pelicula.id_director
     LEFT JOIN 
         premio_mejor_director ON director.id_director = premio_mejor_director.id_director
     GROUP BY 
         director.id_director) AS sub_director ON director.id_director = sub_director.id_director
WHERE 
    premio_mejor_actor.id_premio IS NULL
GROUP BY 
    actor.nombre, pelicula.nombre, director.nombre, sub_director.total_peliculas, sub_director.total_premios; 
```
